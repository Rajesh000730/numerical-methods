```python
 from sympy import *
 import numpy as np
 x, y, h, k = symbols('x y h k', real=True)
 xv = 3.157
 yv = 2.456
f = x**2 - y**2 - 4
g = x**2 + y**2 - 16
fr = f.subs({x:xv, y:yv})
gr = g.subs({x:xv, y:yv})
print(fr, gr)

fx = h*diff(f, x)
fy = k*diff(f, y)
gx = h*diff(g, x)
gy = k*diff(g, y)

hk1 = fx.doit().subs({x:xv, y:yv}) + fy.doit().subs({x:xv, y:yv})
hk2 = gx.doit().subs({x:xv, y:yv}) + gy.doit().subs({x:xv, y:yv})


print(hk1)
print(hk2)

```
